```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df
```

```python
df.isna().sum()
```

```python
df = df.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender']) # Female = 0 | Male = 1
df['Geography'] = le.fit_transform(df['Geography'])
df
```

```python
df['Geography'].unique()
```

```python
df1 = df.drop(columns =['RowNumber','Surname','CustomerId'],inplace =True)
df
```

```python
X= df.drop(columns ='Exited')
Y =df[['Exited']]

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,random_state =42,test_size =0.3)
X_train
```

```python
# APPLYING LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
clf = lr.fit(X_train,Y_train) 
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
acc = accuracy_score(Y_test,Y_pred)
cfr = classification_report(Y_test,Y_pred)
cm = confusion_matrix(Y_test,Y_pred)
print("accuracy is : ",acc*100)
print("classification report is \n",cfr)
print("confusion matrix is  \n",cm)
```

```python
# RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
rf =RandomForestClassifier()
model = rf.fit(X_train,Y_train)
Y1_pred = model.predict(X_test)

acc1 = accuracy_score(Y_test,Y_pred)
cfr1 = classification_report(Y_test,Y_pred)
cm1 = confusion_matrix(Y_test,Y_pred)
print("accuracy is : ",acc1*100)
print("classification report is \n",cfr1)
print("confusion matrix is  \n",cm1)
```

```python
# GRADIENT BOOSTING 
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
mdl = gbc.fit(X_train,Y_train)
y_pred = mdl.predict(X_test)

acc2 = accuracy_score(Y_test,y_pred)
cfr2 = classification_report(Y_test,y_pred)
cm2 = confusion_matrix(Y_test,y_pred)
print("accuracy is : ",acc2*100)
print("classification report is \n",cfr2)
print("confusion matrix is  \n",cm2)
```
